```python
#Estad√≠sticas del Jugador
nombreJugador = nombreUsuario+"1751099"
vidaJugador = 200
nanoescudoJugador = 100
killsJugador = 5
monedasJugador = 1000
puntosEXP = 8256

#Estad√≠sticas del Arma
nombreArma = "M4A1"
da√±oArma = 28
balasDisparadas = 9

#Evento - Da√±o al Jugador con M4A1
def da√±oJugador(): #Funci√≥n que calcula el da√±o que recibe el jugador.
  global vidaJugador, nanoescudoJugador #Permite que la funci√≥n modifique los valores almacenados en dichas variables.

  #Nanoescudo
  da√±oAbsorbidoNano = int(da√±oArma) * 0.25; #El Nanoescudo absorbe un 25% del da√±o.
  estadoNanoescudo = nanoescudoJugador - da√±oAbsorbidoNano * balasDisparadas #El Nanoescudo se reduce con cada disparo.

  #Vida
  da√±oAbsorbidoVida = da√±oArma - da√±oAbsorbidoNano #El da√±o que recibe el jugador disminuye al tener el Nanoescudo.
  estadoJugador = vidaJugador - da√±oAbsorbidoVida * balasDisparadas #El da√±o de cada disparo es menor.

  #Limitando el Da√±o Recibido (No se puede recibir m√°s all√° de...)
  estadoJugador = max(estadoJugador, 0) # Si la vida est√° en 0 el jugador muere.
  estadoNanoescudo = max(estadoNanoescudo, 0)

  #Actualizaci√≥n
  vidaJugador = int(estadoJugador) #Actualiza la vida del jugador luego de recibir los disparos.
  nanoescudoJugador = int(estadoNanoescudo) #Actualiza el nanoescudo del jugador luego de recibir los disparos.
  return estadoNanoescudo, estadoJugador

da√±oJugador() #Llamando a la funci√≥n.
print("Vida: " + str(vidaJugador)) #Imprime la cantidad de vida del jugador luego del combate.
print("Nanoescudo: " + str(nanoescudoJugador)) #Imprime la cantidad de nanoescudo que le queda al jugador.
```

```python
#Evento - Ganar una Partida
def calcular_Puntaje(): #Funci√≥n que calcula los puntajes al ganar una partida.
  global monedasJugador, puntosEXP

  #Recompensa por Victoria
  monedasVictoria = 500
  monedasJugador += monedasVictoria #A√±ade a las monedas que ya ten√≠a el jugador.

  #Puntaje
  puntajeKill = 75 #Puntos que otorga cada kill.
  puntajeGanar = 500 #Puntos que otorga ganar una partida.

  puntajeTotalKill = killsJugador * puntajeKill #Puntos totales que otorgan las kills de una partida.
  puntosTotales = puntajeTotalKill + puntajeGanar #Puntos totales al concluir una partida
  puntosEXP += puntosTotales #A√±ade a la EXP que ya ten√≠a el jugador.

  #Actualizaci√≥n
  return monedasVictoria, puntosTotales

monedasDeVictoria, puntosEXPTotales = calcular_Puntaje() #Llamamos a la funci√≥n utilizando variables de reciente creaci√≥n para almacenar los valores.
print("¬°El Jugador " + str(nombreJugador) + " ha realizado un Booyah!") #Imprime un mensaje de felicitaci√≥n.
print("EXP Obtenida: " + str(puntosEXPTotales) + ".") #Imprime la cantidad de experiencia ganada en la partida.
print("Monedas Obtenidas: +" + str(monedasDeVictoria) + ".") #Imprime la cantidad de monedas ganadas en la partida.
print("[...]")
print("EXP Total: " + str(puntosEXP)) #Imprime la EXP que posee el jugador luego de la partida.
print("Monedas Totales: " + str(monedasJugador)) #Imprime las monedas que posee el jugador luego de la partida.
```
